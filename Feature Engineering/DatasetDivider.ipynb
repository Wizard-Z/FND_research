## Main Code

```python
news_types = ['fake', 'political', 'conspiracy',
              'junksci', 'bias', 'unreliable',
              'unknown', 'satire', 'reliable',
              'hate', 'clickbait', 'rumor'
             ]
```

```python
path = '../Datasets/newsSplit/'
```

```python
# Because for loop looked too messy with these.
def news_filter(newsdf):
    
    global once
    
    def header_once(header=False):
        """Helps to add the header once for every iteration.
        """
        # header gets added after each iteration.
        political.to_csv(path + 'political.csv',mode = 'a', header=header, index=False)
        conspiracy.to_csv(path + 'conspiracy.csv',mode = 'a', header=header, index=False)
        junksci.to_csv(path + 'junksci.csv',mode = 'a', header=header, index=False)
        bias.to_csv(path + 'bias.csv',mode = 'a', header=header, index=False)
        unreliable.to_csv(path + 'unreliable.csv',mode = 'a', header=header, index=False)
        unknown.to_csv(path + 'unknown.csv',mode = 'a', header=header, index=False)
        satire.to_csv(path + 'satire.csv',mode = 'a', header=header, index=False)
        reliable.to_csv(path + 'reliable.csv',mode = 'a', header=header, index=False)
        hate.to_csv(path + 'hate.csv',mode = 'a', header=header, index=False)
        clickbait.to_csv(path + 'clickbait.csv',mode = 'a', header=header, index=False)
        rumor.to_csv(path + 'rumor.csv',mode = 'a', header=header, index=False)
        fake.to_csv(path + 'fake.csv',mode = 'a', header=header, index=False)  
    
    fake = newsdf[newsdf['type'].astype(str) == 'fake']
    political  = newsdf[newsdf['type'].astype(str) == 'political']
    conspiracy = newsdf[newsdf['type'].astype(str) == 'conspiracy']
    junksci = newsdf[newsdf['type'].astype(str) == 'junksci']
    bias = newsdf[newsdf['type'].astype(str) == 'bias']
    unreliable = newsdf[newsdf['type'].astype(str) == 'unreliable']
    unknown = newsdf[newsdf['type'].astype(str) == 'unknown']
    satire = newsdf[newsdf['type'].astype(str) == 'satire']
    reliable = newsdf[newsdf['type'].astype(str) == 'reliable']
    hate = newsdf[newsdf['type'].astype(str) == 'hate']
    clickbait = newsdf[newsdf['type'].astype(str) == 'clickbait']
    rumor = newsdf[newsdf['type'].astype(str) == 'rumor']
    
    if once:
        header_once(once)
        once=False
    else:
        header_once()
```

```python
newsreader = pd.read_csv(open('../Datasets/news_cleaned_2018_02_13.csv', 'rU'), chunksize=100000)
```

```python
once = True

for newsdf in newsreader:
    newsdf.drop(columns=['Unnamed: 0', 'id', 'domain', 'url', 'scraped_at',
                       'inserted_at', 'updated_at', 'keywords','meta_keywords',
                       'meta_description', 'tags', 'summary', 'source'],
              inplace=True)
    news_filter(newsdf)
```
