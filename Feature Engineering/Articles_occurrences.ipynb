## Functions

```python

def addto(dic, keY, positions):
    if keY in dic.keys():
        dic[keY].extend(positions)
    else:
        dic[keY] = positions
    return dic
```

```python

def dist_art(content):
    dist_a = {}
    dist_an = {}
    dist_the = {}
    lines = content.split('\n')
    for line in enumerate(lines):
        for word in enumerate(line[1].split()):
            if word[1] == 'a':
                dist_a = addto(dist_a, line[0], [word[0]])
            if word[1] == 'an':
                dist_an = addto(dist_an, line[0], [word[0]])         
            if word[1] == 'the':
                dist_the = addto(dist_the, line[0], [word[0]])
    return(dist_a,dist_an,dist_the)
             
        
                
```

```python
s = df['content'][4]
a,an,the = dist_art(s)
```

```python

print('occurence of a {}. \n\noccurence of an {} \n\noccurence of the {}'.format(a, an, the))
```
